```python
file_path = '/content/drive/MyDrive/DataAnalyst-mini-projects/Data/LOTR 2.csv'

df2 = pd.read_csv(file_path)
df2
```

```python

df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```




*   **inner** ( jointure interne ) â†’ Conserve uniquement les lignes qui ont une correspondance dans les deux DataFrames
* **outer** ( jointure externe complÃ¨te ) â†’ Conserve toutes les lignes   




```python
# how='left' -> jointure Ã  gauche (LEFT JOIN)
# Toutes les lignes du DataFrame de gauche (df1) sont conservÃ©es, mÃªme si elles nâ€™ont pas de correspondance dans df2 ( Si aucune correspondance nâ€™existe â†’ les colonnes provenant de df2 valent NaN )
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
# Chaque ligne de df1 est combinÃ©e avec chaque ligne de df2
# ðŸ‘‰ Le rÃ©sultat contient toutes les combinaisons possibles entre les deux DataFrames
df1.merge(df2, how = 'cross')
```

```python
# Effectuer une jointure externe complÃ¨te (outer join) entre df1 et df2 Ã  partir de la colonne FellowshipID, tout en Ã©vitant les conflits de noms de colonnes grÃ¢ce aux suffixes
# lsuffix='_Left' -> Suffixe ajoutÃ© aux colonnes de df1
# rsuffix='_Right' -> Suffixe ajoutÃ© aux colonnes de df2
df1.join(df2, on = 'FellowshipID', how = 'outer', lsuffix = '_Left',rsuffix = '_Right')
```

```python
df4 = df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), lsuffix = '_Left',rsuffix = '_Right', how = 'outer')
df4
```

```python
pd.concat([df1,df2], join = 'outer', axis = 1)
```
