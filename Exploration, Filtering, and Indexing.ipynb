* Types de corrélation


1.   Positive (r > 0) : Les variables augmentent ou diminuent ensemble (ex : heures travaillées et salaire).
2.  Négative (r < 0) : Une variable augmente quand l'autre diminue (ex : altitude et température).
3.   Nulle (r ≈ 0) : Aucune relation linéaire perceptible (ex : taille et QI).







```python
df.corr(numeric_only=True)
```

```python
# visualiser graphiquement les corrélations entre les variables numériques
# Transformer une matrice de corrélation (chiffres) en une carte visuelle (heatmap)
sns.heatmap(df.corr(numeric_only=True), annot=True)

plt.rcParams['figure.figsize'] = (20,7)

plt.show()
```

```python
# analyser et comparer les continents en calculant des moyennes par continent, puis en les classant selon la population de 2022
df.groupby('Continent').mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)


# .mean() -> Calcule la moyenne de chaque colonne numérique pour chaque continent (Africa, Asia ...)
# .sort_values(by="2022 Population", ascending=False) -> Trie les continents selon la population moyenne en 2022
# => Objctif:Comparer les continents entre eux sur la base de leurs caractéristiques moyennes, en mettant en évidence ceux dont la population moyenne des pays est la plus élevée.
```

```python
# filtrer le DataFrame pour ne garder que les lignes dont la colonne Continent contient le mot "Oceania"
df[df['Continent'].str.contains('Oceania')]
```

```python
# analyser l’évolution moyenne de la population par continent, puis à classer les continents selon leur population moyenne en 2022.
# => Objectif: Comparer les continents entre eux en calculant la population moyenne des pays à différentes années, et mettre en évidence le continent ayant la population moyenne la plus élevée en 2022.
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
# changer l’orientation du DataFrame df2 afin de faciliter l’analyse et surtout la visualisation de l’évolution de la population dans le temps.
df3 = df2.transpose()
df3
```

```python
# visualiser graphiquement l’évolution des populations moyennes par continent au fil du temps sous forme de graphique en lignes.
import matplotlib.pyplot as plt

# Définir la taille du graphique
plt.figure(figsize=(12, 6))

# Tracer les lignes avec df3
ax = df3.plot(
    linewidth=2,        # épaisseur des lignes
    marker='o',         # marqueurs sur les points
    markersize=6        # taille des marqueurs
)

# Ajouter un titre et des labels
plt.title("Évolution de la population moyenne par continent (1970-2022)", fontsize=16)
plt.xlabel("Années", fontsize=12)
plt.ylabel("Population moyenne (en millions)", fontsize=12)

# Ajouter une grille pour mieux lire les valeurs
plt.grid(True, linestyle='--', alpha=0.5)

# Déplacer et styliser la légende
plt.legend(title="Continent", fontsize=10, title_fontsize=12, loc='upper left')

# Afficher le graphique
plt.show()

```

# *III. Indexing*

```python
df = pd.read_csv(r"/content/drive/MyDrive/DataAnalyst-mini-projects/world_population.csv", index_col = "Country")

df
```

```python
# Réinitialiser l’index du DataFrame et à le remplacer par un index numérique standard (0, 1, 2, …)
df.reset_index(inplace=True)
df
```

```python
# Définir la colonne Country comme index principal du DataFrame → Chaque pays devient l’étiquette (label) d’une ligne
# inplace=True → Applique la modification directement sur df ( Aucun nouveau DataFrame n’est créé )
df.set_index('Country', inplace = True)

df
```

```python
# Supprime l’index actuel
# Le remplace par un index numérique (0,1, …) → Évite les conflits si Country ou Continent étaient déjà des index
# 👉 Étape de nettoyage
df.reset_index(inplace = True)
df
```

```python
df.set_index(['Continent','Country'], inplace=True) # Créer un index à deux niveaux hiérarchique (MultiIndex) basé sur : 'Continent','Country' 👉 Organisation hiérarchique des données
df.sort_index() # Trie le DataFrame selon l’index hiérarchique : continents par ordre alphabétique et Country par ordre alph à l’intérieur de chaque continent

```
