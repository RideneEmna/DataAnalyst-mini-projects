```python
import pandas as pd
```

```python
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/MyDrive/DataAnalyst-mini-projects/Data/Flavors.csv'

df = pd.read_csv(file_path)
df
```

```python
# regroupe les lignes du DataFrame selon les valeurs de la colonne "Base Flavor"
group_by_frame = df.groupby('Base Flavor')

# Calcule la moyenne de chaque colonne numérique pour chaque groupe
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Base Flavor').mean(numeric_only=True)
```

```python
# compter le nombre de valeurs présentes dans chaque groupe, après avoir regroupé le DataFrame selon la colonne "Base Flavor"
df.groupby('Base Flavor').count()
```

```python
df.groupby('Base Flavor').sum()
```

```python
# regrouper les données par 'Base Flavor'
# et  calculer plusieurs statistiques en une seule fois sur deux colonnes précises
# .agg({...}) → Permet d’appliquer plusieurs fonctions statistiques Et sur plusieurs colonnes différentes, en une seule instruction
df.groupby('Base Flavor').agg({'Flavor Rating': ['mean','max','count','sum'], 'Texture Rating':['mean','max','count','sum'] })
```

```python
df.groupby(['Base Flavor','Liked']).agg({'Flavor Rating': ['mean','max','count','sum']})
```
